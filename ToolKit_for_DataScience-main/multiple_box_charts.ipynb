```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
 

sns.set(font_scale=1.9)  
sns.set_style("white")
 
fig, axes = plt.subplots(2, 2, figsize=(19, 15))
 
fig.suptitle('Box Charts')
 
iris = sns.load_dataset("iris")

 
sns.boxplot(ax=axes[0, 0], data=iris, x='species', y='petal_width', linewidth=3 ,saturation=0.9)
sns.boxplot(ax=axes[0, 1], data=iris, x='species', y='petal_length', linewidth=3,saturation=0.9)
sns.boxplot(ax=axes[1, 0], data=iris, x='species', y='sepal_length', linewidth=3,saturation=0.9)
sns.boxplot(ax=axes[1, 1], data=iris, x='species', y='petal_width', linewidth=3,saturation=0.9)


fig.tight_layout() 
plt.savefig('boxchart.png')
plt.show()
```
