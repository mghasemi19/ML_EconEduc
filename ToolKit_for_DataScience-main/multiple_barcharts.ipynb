```python
import numpy as np

import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error
from collections import defaultdict
import numpy as np

import random

def MAE(operation=None):


    truth= [random.randint(2,8) for i in range(500)]

    pred= [random.randint(2,8) for i in range(500)]


    max_length= max(truth)+1


    tr=defaultdict(list)   
    pr=defaultdict(list)  


    for i in range(len(truth)):

        for j in range(2,max_length):

            if truth[i]==j:
                tr[j].append(truth[i])
                pr[j].append(pred[i])


    MAE=[] 

    for i in range(2, max_length) : 

       mae= mean_absolute_error(tr[i],pr[i] ) 
       MAE.append(mae)

    #tupleMAE = tuple(MAE)


    return MAE


tupleMAE_sub= MAE('subtraction')
tupleMAE_con= MAE('concatenation')
tupleMAE_avg= MAE('average')
tupleMAE_had= MAE('hadamard')


opr_dict={'subtraction': '⊖', 'concatenation':'⊕', 'average': '⊘', 'hadamard': '⊙'}

Orion=(0.5, 0.6, 0.8, 1.,1.3, 1.7, 2 )
Rigel=(0.1, 0.5, 0.5, 0.5, 0.6,0.7, 0.5 )

n_groups = 7
bar_width = 0.25
opacity = 0.8

index = np.arange(n_groups)

fig, ((ax1, ax2), (ax3, ax4) )= plt.subplots(nrows=2, ncols=2, figsize=(18, 14))
fig.subplots_adjust(hspace=0.3, wspace=0.2)
ax1.grid(linestyle=':')
ax2.grid(linestyle=':')
ax3.grid(linestyle=':')
ax4.grid(linestyle=':')

rects1 = ax1.bar(index, tupleMAE_sub, bar_width,
                alpha=opacity,
                color='orangered',
                label='SDT',align='center',edgecolor='black',linewidth=1)



rects2 = ax1.bar(index + bar_width, Rigel, bar_width,
                alpha=opacity,
                color='darkblue',
                label='Rigel',align='center',edgecolor='black',linewidth=1)

rects3 = ax1.bar(index + bar_width+ bar_width, Orion, bar_width,
                alpha=opacity,
                color='lime',
                label='Orion',align='center',edgecolor='black',linewidth=1)



ax1.set_title('Vector Operation %s'%opr_dict['subtraction'] , fontsize=22)
ax1.set_ylabel('Mean Absolute Error', fontsize=20)
ax1.set_xlabel('Distance', fontsize=20)


my_xticks = ['2','3 ','4 ',  ' 5', '6',' 7',  '8']
ax1.set_xticks(index + bar_width , ('2', '3', '4', '5','6','7','8'), fontsize=18) 
ax1.set_xticklabels(my_xticks)
#ax1.set_yticks(y_values) 
#ax1.set_yticklabels(my_yticks)
#ax1.set_yticks()
ax1.tick_params(axis='both',labelsize=20)



rects1 = ax2.bar(index, tupleMAE_con, bar_width,
                alpha=opacity,
                color='orangered',
                label='SDT',align='center',edgecolor='black',linewidth=1)


rects2 = ax2.bar(index + bar_width, Rigel, bar_width,
                alpha=opacity,
                color='darkblue',
                label='Rigel',align='center',edgecolor='black',linewidth=1)

rects3 = ax2.bar(index + bar_width+ bar_width, Orion, bar_width,
                alpha=opacity,
                color='lime',
                label='Orion',align='center',edgecolor='black',linewidth=1)


ax2.set_title('Vector Operation %s'%opr_dict['concatenation'] , fontsize=22)
ax2.set_ylabel('Mean Absolute Error', fontsize=20)
ax2.set_xlabel('Distance', fontsize=20)

ax2.set_xticks(index + bar_width , ('2', '3', '4', '5','6','7','8'), fontsize=18) 
ax2.set_xticklabels(my_xticks)
#ax2.set_yticks(y_values) 
#ax2.set_yticklabels(my_yticks)
#ax2.set_yticks()
ax2.tick_params(axis='both',labelsize=20)





rects1 = ax3.bar(index, tupleMAE_avg, bar_width,
                alpha=opacity,
                color='orangered',
                label='SDT',align='center',edgecolor='black',linewidth=1)



rects2 = ax3.bar(index + bar_width, Rigel, bar_width,
                alpha=opacity,
                color='darkblue',
                label='Rigel',align='center',edgecolor='black',linewidth=1)

rects3 = ax3.bar(index + bar_width+ bar_width, Orion, bar_width,
                alpha=opacity,
                color='lime',
                label='Orion',align='center',edgecolor='black',linewidth=1)


ax3.set_title( 'Vector Operation %s'%opr_dict['average'] , fontsize=22)
ax3.set_ylabel('Mean Absolute Error', fontsize=20)
ax3.set_xlabel('Distance', fontsize=20)

ax3.set_xticks(index + bar_width , ('2', '3', '4', '5','6','7','8'), fontsize=18) 
ax3.set_xticklabels(my_xticks)
#ax3.set_yticks(y_values) 
#ax3.set_yticklabels(my_yticks)
#ax3.set_yticks()
ax3.tick_params(axis='both',labelsize=20)




rects1 = ax4.bar(index, tupleMAE_had, bar_width,
                alpha=opacity,
                color='orangered',
                label='SDT',align='center',edgecolor='black',linewidth=1)



rects2 = ax4.bar(index + bar_width, Rigel, bar_width,
                alpha=opacity,
                color='darkblue',
                label='Rigel',align='center',edgecolor='black',linewidth=1)

rects3 = ax4.bar(index + bar_width+ bar_width, Orion, bar_width,
                alpha=opacity,
                color='lime',
                label='Orion',align='center',edgecolor='black',linewidth=1)


ax4.set_title('Vector Operation %s'%opr_dict['hadamard'] , fontsize=22)
ax4.set_ylabel('Mean Absolute Error', fontsize=20)
ax4.set_xlabel('Distance', fontsize=20)

ax4.set_xticks(index + bar_width , ('2', '3', '4', '5','6','7','8'), fontsize=18) 
ax4.set_xticklabels(my_xticks)
#ax4.set_yticks(y_values) 
#ax4.set_yticklabels(my_yticks)
#ax4.set_yticks()
ax4.tick_params(axis='both',labelsize=20)



handles, labels = ax1.get_legend_handles_labels()

fig.legend( handles, labels, loc='upper center', bbox_to_anchor=(0.76,0.96),  fontsize=22) #bbox_to_anchor=(0.5, 0.99)


#fig.set_size_inches(5, 4)

fig.tight_layout() 
plt.savefig('bar_chart.png')
plt.show()
             
              
                  
                
```
