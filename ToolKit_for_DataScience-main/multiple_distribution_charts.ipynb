```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


sns.set(font_scale=1.9)
sns.set_style("white")

df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/diamonds.csv')
x1 = df.loc[df.cut=='Ideal', 'depth']
x2 = df.loc[df.cut=='Fair', 'depth']
x3 = df.loc[df.cut=='Good', 'depth']
x4 = x3*0.3

kwargs = dict(hist_kws={'alpha':.8}, kde_kws={'linewidth':3})
 
fig, axes = plt.subplots(2, 2, figsize=(19, 15))
 
fig.suptitle('Disttribution Charts')
 
 
plot1= sns.distplot(a=x1, ax=axes[0, 0], color="red", label="Compact", **kwargs )
plot2=sns.distplot(a=x3, ax=axes[0, 1], color="deeppink", label="minivan", **kwargs)
plot3=sns.distplot(a=x2, ax=axes[1, 0], color="orange", label="SUV", **kwargs)
plot4=sns.distplot(a=x4, ax=axes[1, 1], color="dodgerblue", label="SUZ", **kwargs)

plot1.legend( loc = 2, bbox_to_anchor = (0.1,1))
plot2.legend( loc = 2, bbox_to_anchor = (0.1,1))
plot3.legend( loc = 2, bbox_to_anchor = (0.1,1))
plot4.legend( loc = 2, bbox_to_anchor = (0.1,1))

fig.tight_layout() 
plt.savefig('distchart.png')


plt.show()

             
```

```python
import seaborn as sns
sns.set_style("white")

# Import data
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/diamonds.csv')
x1 = df.loc[df.cut=='Ideal', 'depth']
x2 = df.loc[df.cut=='Fair', 'depth']
x3 = df.loc[df.cut=='Good', 'depth']

# Plot
kwargs = dict(hist_kws={'alpha':.6}, kde_kws={'linewidth':2})

plt.figure(figsize=(10,7), dpi= 80)
sns.distplot(x1, color="dodgerblue", label="Compact", **kwargs)
sns.distplot(x2, color="orange", label="SUV", **kwargs)
sns.distplot(x3, color="deeppink", label="minivan", **kwargs)
plt.xlim(50,75)
plt.legend();
```
